```python
import mysql.connector

user = "root"
senha = "root"
banco_de_dados = "mercado_br"
host = "localhost"

conexao = mysql.connector.connect(
    host=host,
    user=user,
    password=senha,
    database=banco_de_dados
)



cursor = conexao.cursor()

cursor.execute("SHOW TABLES")


criar_cadastro_empresas = '''
CREATE TABLE IF NOT EXISTS cadastro_empresas (
      keyCompany int AUTO_INCREMENT PRIMARY KEY,
  stockCodeCompany varchar(32) NOT NULL,
  nameCompany varchar(64) NOT NULL,
  sectorCodeCompany varchar(32) NOT NULL,
  sectorCompany varchar(256) NOT NULL,
  segmentCompany varchar(256) NOT NULL

    
)
'''

try:
    cursor.execute(criar_cadastro_empresas)
    print("Tabela de cadastro empresas criada com sucesso")
    cursor.execute("SHOW TABLES")

    
except mysql.connector.Error as erro:
    print(f"Erro ao criar tabela cadastro_empresa: {erro}")
    
    
    


inserir_sql = "INSERT INTO cadastro_empresas (nome, email, telefone) VALUES (%s, %s, %s)"
valores = ("Luiz", "luiz@example.com", "123456789")
cursor.execute(inserir_sql, valores)
conexao.commit()





cursor.close()

print("Conex達o bem-sucedida!")
conexao.close()
```
