```python
import requests 
from bs4 import BeautifulSoup
from pprint import pprint 
import pandas as pd
```

```python
url = r"https://pt.wikipedia.org/wiki/Popula%C3%A7%C3%A3o_mundial"
response = requests.get(url)
response

soup = BeautifulSoup(response.text, 'html.parser') # transforma o text num objeto soup
#soup = BeautifulSoup(r"https://pt.wikipedia.org/wiki/Popula%C3%A7%C3%A3o_mundial", 'html.parser') # transforma o text num objeto soup

tabela = soup.find("table")

df = pd.read_html(str(tabela))[0]

display(df)

# for i, r in enumerate(df):
#     print(i)
#     print(r)
#     print(r["Região"])
```

```python
url_vagas = "https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=python&txtLocation="

response = requests.get(url_vagas)

soup = BeautifulSoup(response.text, 'html.parser')
lista_vagas = soup.find_all("li", class_ = "clearfix job-bx wht-shd-bx" )

pprint(lista_vagas)

```

```python
lista_dfs = []

for vaga in lista_vagas:
    #print(vaga)
    empresa = vaga.h3.text.replace(" ", "").strip()
    posicao = vaga.a.text
    job_description = vaga.find('li', class_='job-description__').text.strip()
    local = vaga.find('li', class_='srp-zindex location-tru').get_text().strip()
    data_publicacao = vaga.find('span', class_ = 'sim-posted').span.text
    
    df = pd.DataFrame(
        {'empresa': empresa,
         'posicao': posicao,
         'job_description': job_description,
         'local': local,
         'data_publicacao': data_publicacao,
         },
        index=[0]
    )
    #pprint(df)
    lista_dfs.append(df)



    #print(job_description)
   # print(empresa, posicao,"\n", job_description,"\n", local, "\n", data_publicacao)
    
    #print(vaga.a.text)
    
    
    
tabela_vagas = pd.concat(lista_dfs, ignore_index=True)
tabela_vagas
```

```python
from selenium.webdriver.common.by import By 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.support.ui import WebDriverWait 


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = 'https://www.bcb.gov.br/estabilidadefinanceira/historicocotacoes'

driver.get(url)


iframe = WebDriverWait(driver, 10).until(
   EC.frame_to_be_available_and_switch_to_it((By.XPATH, "/html/body/iframe"))
)

print(driver.page_source)


#botao_pesquisar = driver.find_element('xpath','/html/body/div/form/div/input')
# Localizar e clicar no botão
botao_pesquisar = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//input[@type='submit' and @value='Pesquisar' and @class='botao']"))
)
botao_pesquisar.click()
#driver.execute_script("arguments[0].click();", botao_pesquisar)

```
