```python
import pandas as pd
import requests
import os
import zipfile
import time
```

```python
anos = range(2010,2025)
url_base = r"https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/" 
dir_base = os.getcwd()
print("site cvm:",url_base)
# Check the current working directory  
print("current_dir:",os.getcwd())  

# Change the current working directory  
os.chdir(r"/home/luiz/cursos/varos/CODIGO_PY/PRINCIPAL/dados_cvm")
#/home/luiz/cursos/varos/CODIGO_PY/PRINCIPAL/dados_cvm
print("current_dir:",os.getcwd())
#"/DFP/DADOS/ "

#os.chdir(dir_base)
#print("current_dir:",os.getcwd())



```

```python
for ano in anos:
    download = requests.get(url_base +  f"DFP/DADOS/dfp_cia_aberta_{ano}.zip")
    open(f"dfp_cia_aberta_{ano}.zip", "wb").write(download.content)
```

```python
lista_demonstracoes_2010_2024 = []
```

```python
diretorio_atual = os.getcwd()
for arquivo in os.listdir(diretorio_atual):
    if(".zip" in arquivo):
        arquivo_zip = zipfile.ZipFile(arquivo)
        tipo = "dfp"
        ano = arquivo[-8:-4]
        #print(ano)
        for planilha in arquivo_zip.namelist():
            #print(planilha)
            cia_aberta = f"{tipo}_cia_aberta_{ano}.csv"
            #print("cia aberta:", cia_aberta)
            if(planilha != cia_aberta ):

                demonstracao = pd.read_csv(arquivo_zip.open(planilha), sep=";", decimal = ",", encoding="ISO=8859-1", chunksize=10000 )
                demonstracao = pd.concat(demonstracao, ignore_index=True)
                demonstracao["tipo_doc"] = len(demonstracao)*[tipo]
                lista_demonstracoes_2010_2024.append(demonstracao)
lista_demonstracoes_2010_2024
        
    
```
