```python
### Merge x Join

dados_roe = pd.DataFrame({"ticker": ['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'MGLU3'],
                         "ROE": [20, 9, 19, 3, 5]})

dados_roic = pd.DataFrame({"codigo": ['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'VIIA3'],
                         "ROIC": [25, 6, 13, 1, 2]})

indicadores = pd.merge(dados_roe, dados_roic, how = "inner", 
               left_on = ['ticker'], right_on = ['codigo'])

indicadores
```

```python
#gabarito 76

mglu = yf.download('MGLU3.SA')
mglu['empresa'] = ['MGLU3'] * len(mglu)

via = yf.download('VIIA3.SA')
via['empresa'] = ['VIIA3'] * len(via)

dados_varejo = pd.concat([mglu, via])
dados_varejo
```

```python
#gabarito 77

from datetime import datetime
from datetime import timedelta

acoes = ['MGLU3.SA', 'CASH3.SA', 'AGRO3.SA']

dia = datetime.now() - timedelta(days = 1)

dados_cotacoes = yf.download(acoes, dia)['Close']
dados_cotacoes = pd.melt(dados_cotacoes, var_name= "cod", value_name= "preco")

dados_volume = yf.download(acoes, dia)['Volume']
dados_volume = pd.melt(dados_volume, var_name= "cod", value_name= "volume")

dados_finais = pd.merge(dados_cotacoes, dados_volume, on = "cod")

dados_finais
```
