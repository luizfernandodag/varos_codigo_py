### Formatação de dados

```python
import pandas as pd
import yfinance as yf
```

```python
acoes = yf.download(['WEGE3.SA', 'VALE3.SA', 'PCAR3.SA', 'LREN3.SA', 
                            'PETR4.SA', 'EQTL3.SA', 'EGIE3.SA', 'ELET3.SA'])['Adj Close']

acoes
```

# Problemas
<BR>

* Base de dados - Cada coluna deve ter uma única variável. 4096 -> limite de colunas MySQL
* Agrupamaneto - É impossível agrupar dados de uma única empresa utilizando groupby.
* Filtragem - Filtrar uma única empresa é possível através de colunas, não por método de filtragem. (Isso também vale pro SQL)

```python
acoes = acoes.reset_index()

acoes
```

```python
#melt transforma um dataframe largo em um dataframe cumprido.
#Cada coluna tem uma única categoria: data, empresa e preço. 

dados_no_melzinho = pd.melt(acoes, id_vars = 'Date', var_name = 'empresa', value_name = 'price')
dados_no_melzinho
```

### Comando SQL

SELECT * FROM dados_cotacoes WHERE empresa = "WEGE3.SA"

```python
#fechamento das empresas

dados_no_melzinho.groupby('empresa')['price'].last()
```

```python
#preço médio

dados_no_melzinho.groupby('empresa')['price'].mean()
```

```python
#filtrar dados

dados_no_melzinho[dados_no_melzinho['empresa'] == "WEGE3.SA"]
```

```python
dados_no_melzinho[(dados_no_melzinho['empresa'] == "WEGE3.SA") |
                 (dados_no_melzinho['empresa'] == "VALE3.SA")]
```

```python
#Operação inversa: pivot table

dados_pivotados = pd.pivot_table(dados_no_melzinho, index='Date', columns = 'empresa')
dados_pivotados
```

```python
dados_pivotados = dados_pivotados.droplevel(level = 0, axis = 1)
```

```python
dados_pivotados
```

```python
#os dados assim são ótimos pra aplicar funções de séries temporais

dados_pivotados.pct_change()
```
